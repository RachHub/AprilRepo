```python
# Get lists of the UUIDs (unique IDs for each EPD) associated with each database
# So I need to do a search for all of the uuids associated with e.g. German database
# First get list of all records in the German database. I am asking the API to return data in JSON format as that's handier for me to manipulate in Python.
api_url = "https://ibudata.lca-data.com/resource/processes?search=true&format=JSON&pageSize=2731&startIndex=0"
response = requests.get(api_url, data=data, headers=headers)
```

```python
# This gives a list of all the EPDs in the German database
print(response)
```

```python
# Read and load the JSON
mydata=json.loads(response.content)
```

```python
# Take out metadata, leaving just EPDs
justdata=mydata['data']
print(justdata)
```

```python
# Now we have to take all the uuids out of the German database and store them
# Get the uuids for German database out of above list
uuids_de =[]
for i in justdata:
  uuids_de.append(i['uuid'])

print(uuids_de)
```

```python
# Now can we get all the German EPDs into this notebook
# We need to loop over all the uuids and call each (this takes 42 mins)
EPDs_de=[]
for i in uuids_de:
  api_url = "https://ibudata.lca-data.com/resource/processes/" + i + "?format=JSON&view=extended"
  response = requests.get(api_url, data=data, headers=headers)
  jsonresponse=json.loads(response.content)
  EPDs_de.append(jsonresponse)
print(EPDs_de[:5])
```
